```python
from keras_vggface.vggface import VGGFace
from keras.models import Model
from keras.layers import Dense

# Load VGGFace model (vgg16 version) without the top classification layer
base_model = VGGFace(model='vgg16' ,pooling='avg')

from keras_vggface.utils import preprocess_input
#from keras.preprocessing import image # This line caused error since image will overwrites the existing object with same name
import numpy as np
from scipy.spatial.distance import cosine
from tensorflow.keras.utils import load_img # Updated import line for load_img (this would work for tf.keras)
from tensorflow.keras.utils import img_to_array # Updated import line for img_to_array (this would work for tf.keras)
# or if your keras version is not tensorflow.keras
#from keras.utils import load_img # Updated import line for load_img (this would work for standalone keras)
#from keras.utils import img_to_array # Updated import line for img_to_array (this would work for standalone keras)

# Function to load and preprocess an image
def load_and_preprocess_img(img_path):
    img = load_img(img_path, target_size=(224, 224)) # load_img is now called correctly.
    img = img_to_array(img) # img_to_array is now called correctly.
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)
    return img

# Load your images (replace with your file names)
img1 = load_and_preprocess_img('emma watson.jpg')  # Same person, image 1
img2 = load_and_preprocess_img('emma watson1.jpg')  # Same person, image 2
img3 = load_and_preprocess_img('train.jpeg')    # Different person

# Extract features using VGGFace
feat1 = base_model.predict(img1)
feat2 = base_model.predict(img2)
feat3 = base_model.predict(img3)

# Compute similarity (higher = more similar)
similarity_same = 1 - cosine(feat1[0], feat2[0])  # Same person
similarity_diff = 1 - cosine(feat1[0], feat3[0])  # Different person

print(f"Similarity between same person: {similarity_same}")
print(f"Similarity between different persons: {similarity_diff}")
```

```python
# Example of face detection with a vggface2 model
from numpy import expand_dims
from matplotlib import pyplot
from PIL import Image
from numpy import asarray
from mtcnn.mtcnn import MTCNN
from keras_vggface.vggface import VGGFace
from keras_vggface.utils import preprocess_input
from keras_vggface.utils import decode_predictions

# extract a single face from a given photograph
def extract_face(filename, required_size=(224, 224)):
	# load image from file
	pixels = pyplot.imread(filename)
	# create the detector, using default weights
	detector = MTCNN()
	# detect faces in the image
	results = detector.detect_faces(pixels)
	# extract the bounding box from the first face
	x1, y1, width, height = results[0]['box']
	x2, y2 = x1 + width, y1 + height
	# extract the face
	face = pixels[y1:y2, x1:x2]
	# resize pixels to the model size
	image = Image.fromarray(face)
	image = image.resize(required_size)
	face_array = asarray(image)
	return face_array

# load the photo and extract the face
pixels = extract_face('train.jpeg')
# convert one face into samples
pixels = pixels.astype('float32')
samples = expand_dims(pixels, axis=0)
# prepare the face for the model, e.g. center pixels
samples = preprocess_input(samples, version=2)
# create a vggface model
model = VGGFace(model='resnet50')
# perform prediction
yhat = model.predict(samples)
# convert prediction into names
results = decode_predictions(yhat)
# display most likely results
for result in results[0]:
	print('%s: %.3f%%' % (result[0], result[1]*100))
```

```python

# face verification with the VGGFace2 model
from matplotlib import pyplot
from PIL import Image
from numpy import asarray
from scipy.spatial.distance import cosine
from mtcnn.mtcnn import MTCNN
from keras_vggface.vggface import VGGFace
from keras_vggface.utils import preprocess_input

# extract a single face from a given photograph
def extract_face(filename, required_size=(224, 224)):
	# load image from file
	pixels = pyplot.imread(filename)
	# create the detector, using default weights
	detector = MTCNN()
	# detect faces in the image
	results = detector.detect_faces(pixels)
	# extract the bounding box from the first face
	x1, y1, width, height = results[0]['box']
	x2, y2 = x1 + width, y1 + height
	# extract the face
	face = pixels[y1:y2, x1:x2]
	# resize pixels to the model size
	image = Image.fromarray(face)
	image = image.resize(required_size)
	face_array = asarray(image)
	return face_array

# extract faces and calculate face embeddings for a list of photo files
def get_embeddings(filenames):
	# extract faces
	faces = [extract_face(f) for f in filenames]
	# convert into an array of samples
	samples = asarray(faces, 'float32')
	# prepare the face for the model, e.g. center pixels
	samples = preprocess_input(samples, version=2)
	# create a vggface model
	model = VGGFace(model='resnet50', include_top=False, input_shape=(224, 224, 3), pooling='avg')
	# perform prediction
	yhat = model.predict(samples)
	return yhat

# determine if a candidate face is a match for a known face
def is_match(known_embedding, candidate_embedding, thresh=0.5):
	# calculate distance between embeddings
	score = cosine(known_embedding, candidate_embedding)
	if score <= thresh:
		print('>face is a Match (%.3f <= %.3f)' % (score, thresh))
	else:
		print('>face is NOT a Match (%.3f > %.3f)' % (score, thresh))

# define filenames
filenames = ['train.jpeg', 'babu2.jpeg',
	'babu.jpeg', 'emma watson.jpg']
# get embeddings file filenames
embeddings = get_embeddings(filenames)
# define praharsh
praharsh_id = embeddings[0]
# verify known photos of praharsh
print('Positive Tests')
is_match(embeddings[0], embeddings[1])
is_match(embeddings[0], embeddings[2])
# verify known photos of other people
print('Negative Tests')
is_match(embeddings[0], embeddings[3])
```
